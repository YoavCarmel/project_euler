```python
from tqdm import trange
from typing import List, Tuple
from libs.calculations import diffs_list
from math import log
from libs.numbers_properties import is_int
```

```python
def get_seq(target, nums):
    ll = list()
    for num in nums:
        ll.append(target // num)
        target %= num
    return ll
```

```python
def is_p12n(num):
    return str(num)[:2] == "12"
```

```python
def get_12_diffs():
    num = 1
    powers = list()
    for p in trange(1, 10 ** 4):
        num *= 2
        if is_p12n(num):
            powers.append(p)
    return diffs_list(powers)
```

```python
set(get_12_diffs())
```

```python
def is_p123n(num):
    return str(num)[:3] == "123"
```

```python
def get_123_diffs():
    res = 2 ** 80
    power = 80
    checked_diffs = {d: 2 ** d for d in {10, 73, 83}}
    diffs_keys_sorted = sorted(checked_diffs.keys())
    powers = list()
    for _ in trange(1, 3000):
        for diff in diffs_keys_sorted:
            new_res = res * checked_diffs[diff]
            if is_p12n(new_res):
                power += diff
                res = new_res
                if is_p123n(res):
                    powers.append(power)
                break
        else:
            print(power, powers[-1])
    return diffs_list(powers)
```

```python
diffs123 = get_123_diffs()
```

```python
set(diffs123)
```

```python
def p_123_1() -> Tuple[int, int]:
    num = 1
    p = 1
    while True:
        num *= 2
        if is_p123n(num):
            return p, num
        p += 1
```

```python
def first_123_powers():
    power, num = p_123_1()
    checked_diffs = {d: 2 ** d for d in {196, 289, 485}}
    diffs_keys_sorted = sorted(checked_diffs.keys())
    powers = list()
    for _ in trange(1, 500):
        for diff in diffs_keys_sorted:
            new_res = num * checked_diffs[diff]
            if is_p123n(new_res):
                power += diff
                num = new_res
                powers.append(power)
                break
        else:
            print(power, powers[-1])
            break
    return powers
```

```python
diffs123 = diffs_list(first_123_powers())
```

```python
def values_indices(lst):
    values = {v: list() for v in set(lst)}
    for i, num in enumerate(lst):
        values[num].append(i)
    return values
```

```python
def print_special_seq(seq, new_line_num):
    for i, v in enumerate(seq[:-1]):
        if seq[i] == new_line_num:
            print(seq[i], end=", ")
        elif seq[i + 1] == new_line_num:
            print(seq[i])
        else:
            print(seq[i], end=", ")

```

```python
def powers_rounding(last_power):
    count = 1
    n = 90
    while count < last_power:
        for jump in [196, 289, 485]:
            temp_n = n + jump
            c = temp_n * log(2, 10) - log(1.23, 10)
            if abs(c - int(c)) < log(1.24, 10) - log(1.23, 10):
                count += 1
                n += jump
                break
    return n
```
